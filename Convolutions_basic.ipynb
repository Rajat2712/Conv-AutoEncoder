```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
im = cv2.imread('./img.jpg')
img = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
plt.show()
```

```python
K = np.array([[-1, -1, -1],
              [-1,  8, -1],
              [-1, -1, -1]])
print K
```

```python
out_r = img.shape[0] - K.shape[0] + 1
out_c = img.shape[1] - K.shape[1] + 1

X = np.zeros((out_r, out_c))
print img.shape, K.shape, X.shape
```

## Convolution

```python
for rx in range(out_r):
    for cx in range(out_c):
        image_patch = img[rx:K.shape[0]+rx, cx:K.shape[1]+cx]
        prod = image_patch * K
        # Compute the value to be put at this pixel
        X[rx, cx] = prod.sum()
```

```python
X_ = X * (X > 0)
plt.imshow(X_, cmap='gray')

plt.show()
```
